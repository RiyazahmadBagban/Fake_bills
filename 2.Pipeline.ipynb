```python
import pandas as pd

```

```python
dataset = pd.read_csv('fake_bills_proper_clean_data.csv')
dataset
```

```python
dataset = dataset.drop(["Unnamed: 0"],axis=1)
dataset
```

```python
dataset.describe()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
```

```python
classifier = []
# classifier.append(("mapper",m))
classifier.append(('standardize', StandardScaler()))
classifier.append(("model",GaussianNB()))
```

```python
X = dataset.drop(columns = 'is_genuine')
y = dataset['is_genuine'].values
```

```python
model = Pipeline(classifier)
model.fit(X,y)
```
